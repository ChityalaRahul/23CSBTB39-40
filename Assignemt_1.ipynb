```python
import numpy as np
import pandas as pd
import requests
from google.colab import files

# Create lists
list1 = [1, 2, 3, 4, 5]
list2 = [6, 7, 8, 9, 10]

# Print lists and their types
print("List1:", list1)
print("Type of List1:", type(list1))
print("List2:", list2)
print("Type of List2:", type(list2))
# Convert lists to numpy arrays
array1 = np.array(list1)
array2 = np.array(list2)

# Print arrays and their types
print("Array1:", array1)
print("Type of Array1:", type(array1))
print("Array2:", array2)
print("Type of Array2:", type(array2))
# Already identified above
# Mathematical operations
addition = array1 + array2
multiplication = array1 * array2
division = array1 / array2
power = np.power(array1, 2)

print("Addition:\n", addition)
print("Multiplication:\n", multiplication)
print("Division:\n", division)
print("Power:\n", power)
# Generating textual output
textual_output = f"Addition of Two:\n{array1} + {array2} = {addition}"
print(textual_output)
# Implementing numpy functions
sin_values = np.sin(array1)
log_values = np.log(array1 + 1)  # Adding 1 to avoid log(0) issue
log2_values = np.log2(array1 + 1)
exp_values = np.exp(array1)

print("Sine values:\n", sin_values)
print("Log values:\n", log_values)
print("Log2 values:\n", log2_values)
print("Exponential values:\n", exp_values)
import matplotlib.pyplot as plt

# Data
people = ['kiran', 'arun', 'vijay', 'varun']
age = [25, 30, 35, 40]
height = [145, 151, 165, 173]
weight = [45, 55, 65, 75]

plt.figure(figsize=(15, 10))

# Scatter plot of height vs age
plt.subplot(2, 2, 1)
plt.scatter(age, height, color='blue')
plt.title('Scatter Plot of Height vs Age')
plt.xlabel('Age')
plt.ylabel('Height (cm)')
plt.grid(True)
# Bar chart of weight
plt.subplot(2, 2, 2)
plt.bar(people, weight, color='green')
plt.title('Bar Chart of Weight by Person')
plt.xlabel('Person')
plt.ylabel('Weight (kg)')
plt.grid(True)
# Histogram of age
plt.subplot(2, 2, 3)
plt.hist(age, bins=4, color='orange', edgecolor='black')
plt.title('Histogram of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)

plt.tight_layout()
plt.show()
#part-3
uploaded = files.upload()
# Read the uploaded file
df = pd.read_csv("heart.csv")
print(df.head())
print(df)
# Read the uploaded tsv file
#df = pd.read_csv("number.tsv", sep='\t', on_bad_lines='skip')
#print(df.head())
#print(df)
```
